```python
import pandas
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter


df = pandas.read_csv('john_hopkins_brac_nsu_research.csv',encoding="ISO-8859-1",parse_dates=True)
df['Date'] = pandas.to_datetime(df['Date'], format = '%d/%m/%Y') # convert date


ax1 = plt.subplot()
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax1.xaxis.set_minor_locator(mdates.DayLocator(interval=1))
ax1.xaxis.set_major_locator(mdates.DayLocator(interval=5))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
ax1.plot_date(df['Date'].iloc[1:118], df['Death'].iloc[1:118], '-', label = 'Death')
ax1.plot_date(df['Date'].iloc[2:97], df['New_Cases'].iloc[2:97], '-', label = 'New_Cases')
ax1.plot_date(df['Date'].iloc[2:103], df['Symtotic_Cases'].iloc[2:103], '-', label = 'Symtotic_Cases')
ax1.plot_date(df['Date'].iloc[2:116], df['Cases Needing Crittical Care'].iloc[2:116], '-', label = 'Cases Needing Crittical Care')
ax1.plot_date(df['Date'].iloc[2:108], df['Cases_Needing_Hospitalization'].iloc[2:108], '-', label = 'Cases_Needing_Hospitalization')
ax1.legend()
plt.gcf().set_size_inches(18,13)
plt.xticks(rotation=90) 
plt.gcf().tight_layout() 


# issue: https://github.com/matplotlib/matplotlib/issues/9610
df.set_index('Date', inplace = True)
df.index.to_pydatetime()
plt.show()
```
