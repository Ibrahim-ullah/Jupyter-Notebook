```python
import pandas as pd
import matplotlib.pyplot as plt

confirmed = pd.read_csv('confirmed.csv')
recovered= pd.read_csv('recovered.csv')
deaths = pd.read_csv('deaths.csv')
confirmed.set_index('Country/Region', inplace=True)
recovered.set_index('Country/Region', inplace=True)
deaths.set_index('Country/Region', inplace=True)
```

```python
subset_confirmed = confirmed.loc['Bangladesh']
subset_recovered = recovered.loc['Bangladesh']
subset_deaths = deaths.loc['Bangladesh']
```

```python
a = subset_confirmed.loc['4/1/2020':'4/16/2020']
b = subset_recovered.loc['4/1/2020':'4/16/2020']
c = subset_deaths.loc['4/1/2020':'4/16/2020']
```

```python
plt.xticks(rotation=90)
plt.gcf().set_size_inches(18,13)
plt.plot(a.index, a.values, b.index, b.values, c.index, c.values)
plt.show()
```

```python
ax = plt.subplot(111)
ax.plot(a.index, a.values, label='Infected')
ax.plot(b.index, b.values, label='Recovered')
ax.plot(c.index, c.values, label='Deaths')
ax.legend()
plt.xticks(rotation=90)
plt.gcf().set_size_inches(18,8)
plt.show()
```
